### Import Statments

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy import stats
import seaborn as sns
```

### Defining and Merging Datasets

```python
# column names
col = ["HCHO reading", "Location", "Current Date", "Next Date"]

# Add column names to CSV files
d1 = pd.read_csv("col_mat_nuw_output.csv", names=col)
d2 = pd.read_csv("mon_kur_jaf_output.csv", names=col)
d3 = pd.read_csv("kan_output.csv", names=col)

# Merge datasets
merged = pd.concat([d1, d2, d3])
merged
```

### Filling nulls and handling outliers

```python
print("missing values before filling:", merged.isnull().sum().sum())

# Fill missing values using forward fill
merged.fillna(method='ffill', inplace=True)

# Remove outliers from  'HCHO reading' 
def remove_outliers(df, column):
    z = stats.zscore(df[column])
    z_scores = abs(z)
    filtered = (z_scores < 3)
    return df[filtered]

merged = remove_outliers(merged, 'HCHO reading')

# Make negative values positive
merged['HCHO reading'] = merged['HCHO reading'].abs()
print("missing values after filling:", merged.isnull().sum().sum())

```

### Plot for HCHO levels in each city

```python
plt.figure(figsize=(12, 6))

cities = merged['Location'].unique()

for city in cities:
    data = merged[merged['Location'] == city].copy()  
    # Convert to datetime
    data['Current Date'] = pd.to_datetime(data['Current Date'])
    # Extract year
    data['Year'] = data['Current Date'].dt.year
    mean_hcho = data.groupby('Year')['HCHO reading'].mean()
    plt.plot(mean_hcho.index, mean_hcho.values, label=city)

plt.xlabel('Year')
plt.ylabel('Mean HCHO Levels')
plt.title('Mean HCHO Levels of cities')
plt.legend(loc="upper right")

# Rotate x-axis tick labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

### BoxPlot for for HCHO levels in each city


```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Location', y='HCHO reading', data=merged, color='pink')  # Set color to skyblue
plt.title('Boxplot of HCHO Levels for Each City')
plt.xlabel('City')
plt.ylabel('HCHO Levels')
plt.xticks(rotation=45)
plt.show()
```

### Mean, median, and standard deviation of HCHO levels 

```python
city_stats = merged.groupby('Location')['HCHO reading'].agg(['mean', 'median', 'std'])
overall_stats = merged['HCHO reading'].agg(['mean', 'median', 'std'])

print(city_stats)
print("\n----------Statistics----------")
print(overall_stats)
```

### Read new dataset


```python
#View unique values
weather_vals = pd.read_csv(r"C:\Users\gaind\Desktop\Data_Engineering\3662070.csv")
print (weather_vals["NAME"].unique())
```

### Remove unique "NAME" values

```python
weather = weather_vals[~weather_vals['NAME'].str.contains('puttalam', case=False) & ~weather_vals['NAME'].str.contains('china bay', case=False)]
weather
```

### Remove "STATION" column

```python
#Drop stations coloumn
weather = weather.drop(columns = ["STATION"])
weather
```

### View missing values

```python
weather.isnull().sum()
```

### Fill null values using mean

```python
# Calculate the mean 
mean = weather[['PRCP', 'TMAX', 'TMIN']].mean()

#Fill null values
weather[['PRCP', 'TMAX', 'TMIN']] = weather[['PRCP', 'TMAX', 'TMIN']].fillna(mean)
print("\nMissing values after filling")
print(weather[['PRCP', 'TMAX', 'TMIN']].isnull().sum())
```

### Change coloumn names 

```python
# name to location , Date to current date

weather.rename(columns={'NAME': 'Location', 'DATE': 'Current Date'}, inplace=True)
weather.columns
```

### Making city names similar to ones in previous dataset


```python
# Replace values in the 'Location' 
replace_dict = {
    'COLOMBO, CE': 'Colombo proper',
    'KURUNEGALA, CE': 'Kurunegala proper',
    'NUWARA ELIYA, CE': 'Nuwara Eliya proper'
}

weather['Location'] = weather['Location'].map(replace_dict).fillna(weather['Location'])
print(weather['Location'].unique())
```
