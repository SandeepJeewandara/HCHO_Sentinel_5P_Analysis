```python
# Initialize SparkSession
spark = SparkSession.builder.appName('HCHO_Analysis').getOrCreate()
```

#### Convert Data Types

```python
# Convert Current_Date column to Date Type
data = data.withColumn('Current_Date', to_date(col('Current_Date'), 'yyyy-MM-dd'))

# Convert "HCHO_reading" column to numeric
data = data.withColumn('HCHO_reading', col('HCHO_reading').cast('double'))
```

```python
# Explore missing values in each column
for column in data.columns:
    missing_count =  data.filter(col(column).isNull()).count()
    print(f"Missing values in column '{column}': {missing_count}")
```

```python
# Define the window
windowSpec = Window.orderBy().rowsBetween(Window.unboundedPreceding, 0)

# Fill null values with Foward fill method
data = data.withColumn('HCHO_reading', last('HCHO_reading', True).over(windowSpec))
```

## Data Visualization

```python
# Convert Spark DataFrame to Pandas DataFrame  for Plotting
data_pd = data.toPandas()

# Convert "Current_Date" Column to Datetime Type
data_pd['Current_Date'] = pd.to_datetime(data_pd['Current_Date'])
```

```python
# Function For Data Visualization

def HCHO_levels(location):
    # Filter data for location
    location_data = data_pd[data_pd['Location'] == location]

    # Extract year and month 
    location_data['Year'] = location_data['Current_Date'].dt.year
    location_data['Month'] = location_data['Current_Date'].dt.month

    # Create a new column combining year and month
    location_data['Year_Month'] = location_data['Current_Date'].dt.to_period('M')

    # Group data by year-month and calculate the mean HCHO reading for each month
    location_monthly_mean = location_data.groupby('Year_Month')['HCHO_reading'].mean()

    # Convert index to string for plotting
    location_monthly_mean.index = location_monthly_mean.index.strftime('%Y-%m')

    # Plotting
    plt.figure(figsize=(20, 6))
    
    # Create plot
    plt.plot(location_monthly_mean.index, location_monthly_mean, marker='.', linestyle='--', label=location)

    # Add labels and title
    plt.title(f'HCHO Level Changes in {location} Over Time')
    plt.xlabel('Year-Month')
    plt.ylabel('Mean HCHO Reading')
    plt.grid(True)
    plt.xticks(rotation=90)
    plt.show()
```

## Address Outliers

#### Manage Outliers Using IQR Method

```python
# Calculate quartiles
quantiles = data.approxQuantile("HCHO_reading", [0.25, 0.75], 0.05) 

print(quantiles)
    
# Calculate IQR
q1 = quantiles[0]
q3 = quantiles[1]
iqr = q3 - q1

# Define lower and upper bounds
lower= q1 - 1.5 * iqr
upper= q3 + 1.5 * iqr

# Filter out outliers
final_data = data.filter((F.col('HCHO_reading') >= lower) & (F.col('HCHO_reading') <= upper))
```

#### Convert Preprocessed Data to Pandas Dataframe

```python
# Convert filtered data to Pandas DataFrame for plotting
final_pd = final_data.toPandas()


# Convert "Current_Date" column to datetime
final_pd['Current_Date'] = pd.to_datetime(final_pd['Current_Date'])
    
# Set "Current_Date" as index
#final_pd.set_index('Current_Date', inplace=True)
```

```python
from datetime import datetime
import matplotlib.dates as mdates

def covid_impact(location):
    # COVID-19 lockdown Period
    lockdown_start_date = datetime.strptime("2020-03-15", "%Y-%m-%d")
    lockdown_end_date = datetime.strptime("2020-05-15", "%Y-%m-%d")

    # Filter data for location
    location_data = final_pd[final_pd['Location'] == location]

    # Extract year and month 
    location_data['Year'] = location_data['Current_Date'].dt.year
    location_data['Month'] = location_data['Current_Date'].dt.month

    # Create a new column combining year and month
    location_data['Year_Month'] = location_data['Current_Date'].dt.to_period('M')

    # Group data by year-month and calculate the mean HCHO reading for each month
    location_monthly_mean = location_data.groupby('Year_Month')['HCHO_reading'].mean()

    # Convert index to timestamps
    location_monthly_mean.index = location_monthly_mean.index.to_timestamp()

    # Create plot
    plt.figure(figsize=(20, 6))
    plt.plot(location_monthly_mean.index, location_monthly_mean, marker='.', linestyle='--', label=location)

    # Add vertical markers for lockdown start and end dates
    plt.axvline(lockdown_start_date, color='red', linestyle='--', label='Lockdown Start')
    plt.axvline(lockdown_end_date, color='green', linestyle='--', label='Lockdown End')

    # Add labels and title
    plt.title(f'COVID-19 Lockdown Impact Analysis - {location}')
    plt.xlabel('Year')
    plt.ylabel('Mean HCHO Reading')
    plt.grid(True)
    plt.xticks(rotation=90)

    # Show legend
    plt.legend()
    plt.show()
```

## Analyse Trends Over Time

```python
# Import necessary libraries for time series analysis
from statsmodels.tsa.seasonal import seasonal_decompose    
```

```python
# Perform time series analysis for each city
def analyze_trends(location):
    analyze_data = final_pd[final_pd['Location'] == location] 
    
    # Perform seasonal decomposition
    decomposition = seasonal_decompose(analyze_data['HCHO_reading'], model='additive', period=365)
    
    # Extract components
    trend = decomposition.trend
    seasonal = decomposition.seasonal
    residual = decomposition.resid
    
    # Plot components
    plt.figure(figsize=(12, 8))
    
    plt.subplot(411)
    plt.plot(analyze_data['HCHO_reading'], label='Original')
    plt.legend(loc='best')
    
    plt.subplot(412)
    plt.plot(trend, label='Trend')
    plt.legend(loc='best')
    
    plt.subplot(413)
    plt.plot(seasonal, label='Seasonal')
    plt.legend(loc='best')
    
    plt.tight_layout()
    plt.show()
```

```python
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from datetime import datetime
from statsmodels.tools.eval_measures import rmse
from sklearn.model_selection import train_test_split
from datetime import datetime
```

### Arima Model

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
def train_arima_and_predict(data, location):
    # Filter data for the specified location
    location_data = data[data['Location'] == location].copy()
    location_data.index = pd.to_datetime(location_data.index)

    # Split the data into training and testing sets
    test_data = location_data.iloc[-30:]
    train_data = location_data.iloc[:-30]
    
    # Find the best ARIMA parameters using auto_arima
    auto_arima_model = auto_arima(train_data['HCHO_reading'], seasonal=False, trace=True)

    # Fit the ARIMA model
    arima_model = ARIMA(train_data['HCHO_reading'], order=auto_arima_model.order)
    arima_fit = arima_model.fit()

    # Predict HCHO values
    forecast_length = 30
    predicted_values = arima_fit.forecast(steps=forecast_length)

    # Calculate evaluation metrics
    mse = mean_squared_error(test_data['HCHO_reading'], predicted_values)
    r_squared = r2_score(test_data['HCHO_reading'], predicted_values)
    mae = mean_absolute_error(test_data['HCHO_reading'], predicted_values)

    # Print evaluation metrics
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"R-squared (R^2): {r_squared}")
    print(f"Mean Absolute Error (MAE): {mae}")

    # Plot test data and predictions
    plt.figure(figsize=(10, 6))
    plt.plot(test_data.index, test_data['HCHO_reading'], label='Actual')
    plt.plot(test_data.index, predicted_values, color='red', label='Predicted')
    plt.title(f'HCHO Prediction for {location} in December 2023')
    plt.xlabel('Date')
    plt.ylabel('HCHO Reading')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

```python
train_arima_and_predict(model_pd,'Colombo Proper')
```
